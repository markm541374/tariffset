```python
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
import time
plt.ion()
%matplotlib inline
```

```python
def kf1(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return 0.1*sp.exp(-((d[0,0])/0.1))
def kf2(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return 0.2*sp.exp(-((d[0,0])/0.03))
def kf3(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return 0.01*sp.exp(-((d[0,0])/0.001))
def kf4(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return 2.5*sp.exp(-((d[0,0])/0.003))
```

```python

import tariffopt as TO
e=TO.experiment([kf1,kf2,kf3],["ts0.txt","ts1.txt","ts2.txt"],[100,100,100,100,100],[0,0,0,0,0],TO.gen_3interp_tariff)

#e.loadtrace('trace.txt')
```

```python
def sqexpkern(x1,x2,l,A):
    delta=x1-x2
    d=delta*delta.T
    return A*sp.exp(-((d[0,0])/float(l)))

for ll in [0.001,0.003,0.01,0.03,0.1,0.3,1.,3.,10]:
    for Aa in [0.0001,0.0003,0.001,0.003,0.01,0.03,0.1,0.3,1.,3.]:
        kf=lambda x,y:sqexpkern(x,y,ll,Aa)
        llk=e.G.eval_kernel_llk(kf)
        print "l= "+str(ll)+" A= "+str(Aa)+" llk= "+str(llk)
```

```python
def gen_sqexp_k(theta):
	A=theta[0]
	d=sp.matrix(theta[1:]).T
	
	N=sp.matrix(d).shape[0]
	D=sp.eye(N)
	for i in range(N):
		D[i,i]=1./(d[i,0]**2)
	return lambda x,y:A*sp.exp(-0.5*(x-y)*D*(x-y).T)

def m52_k(x,y,A,D):
	r2=(x-y)*D*(x-y).T
	tmp=sp.sqrt(5*r2)
	return A*(1+tmp+5*r2/3.)*sp.exp(-tmp)

def gen_m52_k(theta):
	A=theta[0]
	d=sp.matrix(theta[1:]).T
	
	N=sp.matrix(d).shape[0]
	D=sp.eye(N)
	for i in range(N):
		D[i,i]=1./(d[i,0]**2)
	
	
	return lambda x,y:m52_k(x,y,A,D)

#kf=gen_sqexp_k([0.1,0.075,15.,15.,15.,15.])
kf=gen_sqexp_k([0.95,0.5,0.3])
#def kf4(x1,x2):
#    delta=x1-x2
#    d=delta*delta.T
#    return 0.01*sp.exp(-((d[0,0])/0.03))
```

```python
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
import time
plt.ion()
%matplotlib inline
```

```python
def kf1(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return sp.exp(-((d[0,0])/1.0))
def kf2(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return sp.exp(-((d[0,0])/0.5))
def kf3(x1,x2):
    delta=x1-x2
    d=delta*delta.T
    return sp.exp(-((d[0,0])/0.2))
```

```python
import tariffopt as TO
e=TO.experiment([kf1,kf2,kf3],["ts0.txt","ts1.txt","ts2.txt","ts3.txt","ts4.txt","ts5.txt"],[2,2,2,2,2],[0,0,0,0,0],5)

#e.loadtrace('fname.txt')
```

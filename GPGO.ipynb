```python
import GPy
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
from scipy.stats import norm as norm
import DIRECT
import operator
import time
import sys
plt.ion()
%matplotlib inline
class GPGO():
    def __init__(self,optfn,lower,upper,dim):
        self.dim=dim
        self.optfn=optfn
        self.upper=upper
        self.lower=lower
        self.trace=None
        self.nGPs=3
        self.GPs=[None]*self.nGPs
        self.lls=[None]*self.nGPs
        
        self.best=None
        
        self.noiseval=0.0001
        self.lengthscales=[1.4]
        self.outputscales=[1.]
        
        self.nGPs=len(self.lengthscales)*len(self.outputscales)
        self.kernels=[]
        self.GPs=[None]*self.nGPs
        self.lls=[None]*self.nGPs
        for l in self.lengthscales:
            for o in self.outputscales:
                self.kernels.append(GPy.kern.rbf(input_dim=self.dim, variance=o, lengthscale=l))
        return
    
    def makestep(self,graphs=False, allGraphs=False):
        print "\n--x--x--x--x--"
        print "finding next eval location"
        t0=time.time()
        [x,flag]=self.findEIopt()
        t1=time.time()
        if flag:
            print "optimisation failed"
            return -1
        print "next evaluation: "+str(x)
        print "search time(s): "+str(t1-t0)
        self.evaluate(x)
        if graphs:
            self.plot(plotall=allGraphs)
        t2=time.time()
        print "step time: "+str(t2-t0)
        print "--x--x--x--x--\n"
        
        
    def evaluate(self,x):
        f = self.optfn(x)
        
        print "evaluation returned: "+str(f)
        if self.trace==None:
            self.trace=np.hstack([f,x]).reshape(1,self.dim+1)
            self.best=[f,x,0]
        else:
            self.trace=np.vstack([self.trace,np.hstack([f,x])])
            if self.best[0]>f:
                self.best=[f,x,self.trace.shape[0]-1]
        
        for i,k in enumerate(self.kernels):
            
            self.GPs[i]=GPy.models.GPRegression(self.trace[:,1:],self.trace[:,0:1],k)
            self.GPs[i].constrain_fixed('.*noise',self.noiseval)
            self.lls[i]=self.GPs[i].log_likelihood()
           
        return f
    
    def calcEIpoint(self,ER,mu,sigma):
        alpha=(-ER+mu)/sigma
        Z=norm.cdf(-alpha)
        if Z==0.0:
            return 0.0
        E=-mu+norm.pdf(alpha)*sigma/Z+ER
        EI=(norm.cdf(-alpha))*E
        if np.isfinite(EI):
            return EI
        else:
            print "notfinite"
            return 0.0
        
    def calcEI(self,ER,mu,sigma):
        ER=ER*np.ones(mu.shape) 
        EI=map(self.calcEIpoint,ER,mu,sigma)
        return EI
        
    def singleEI(self,x,usrdata,i):
        [mean,cov,lower,upper]=self.GPs[i].predict(np.array(x), full_cov=False)
        t1=self.calcEI(self.best[0],mean,cov)
        EI=np.array(map(operator.neg,t1)).T
        return mean,cov,EI
    
    def EIoptfn(self,x,usrdata):
        EI=0.0
        for i in range(self.nGPs):
            _,_,EIs=self.singleEI(x,usrdata,i)
            #no idea why but EIs becomes 1d for values of x that are equal to existing ones
            try:
                EI+=EIs[0,0]*sp.exp(self.lls[i])
            except IndexError:
                EI+=EIs[0]*sp.exp(self.lls[i])
        
        return EI,0
    
    def findEIopt(self):
        [x,EImin,ierror]=DIRECT.solve(self.EIoptfn,self.lower,self.upper,user_data=None,algmethod=1,maxf=5000)
        print ierror
        flag=0
        if max(x)==0.0 and min(x)==0.0:
            flag=1
       
        return [x,flag]
    
    def plot(self,plotall=False):
        size=120
        xaxis=np.linspace(-3,3,size)
        yaxis=np.linspace(-3,3,size)
        x=np.ndarray(0)
        y=np.ndarray(0)
        for a in xaxis:
            x=np.hstack([x, [a]*size])
            y=np.hstack([y,yaxis])
        X=np.vstack([x,y])
        
        EII=np.zeros([size**2,1])
        meanI=np.zeros([size**2,1])
        for i in range(self.nGPs):
            mean,cov,EI = self.singleEI(X.T,None,i)
            EI=EI.reshape([size**2,1])
            mean=mean.reshape([size**2,1])
            t1=np.exp(self.lls[i])*EI
            t2=np.exp(self.lls[i])*mean
            EII+=t1
            meanI+=t2 # not sure this is accurate, need to check the base of log likelihood
            mean_grid= np.hstack(split(mean,size))
            cov_grid= np.hstack(split(cov,size))
            EI_grid= np.hstack(split(EI,size))
            if plotall:
                print "\n-x-x-x-x-"
                print "Displaying all hyperparameter options"
                plt.figure(figsize=(20,5))
                plt.subplot(131)
                try:
                    plt.contour(xaxis,yaxis,mean_grid,30)
                except:
                    pass
                plt.subplot(132)
                try:
                    plt.contour(xaxis,yaxis,cov_grid,30)
                except:
                    pass
                plt.subplot(133)
                try:
                    plt.contour(xaxis,yaxis,EI_grid,30)
                except:
                    pass
            
                print "kernel: "+str(self.GPs[i].kern)
                print "EI range: "+str(max(EI))+" --> "+str(min(EI))
                print "llk: " +str(self.lls[i])
                print "-x-x-x-x-\n"
            
        plt.figure(figsize=(20,5))
        plt.subplot(133)
        EII_grid= np.hstack(split(EII,size))
        
        try:
            plt.contour(xaxis,yaxis,EII_grid,30)
        except:
                pass
        plt.subplot(131)
        meanI_grid= np.hstack(split(meanI,size))
        try:
            plt.contour(xaxis,yaxis,meanI_grid,30)
        except:
                pass
        plt.show()
        return
```

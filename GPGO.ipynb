```python
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
import time
plt.ion()
%matplotlib inline
```

```python
def gen_sqexp_k(theta):
	A=theta[0]
	d=sp.matrix(theta[1:]).T
	
	N=sp.matrix(d).shape[0]
	D=sp.eye(N)
	for i in range(N):
		D[i,i]=1./(d[i,0]**2)
	return lambda x,y:A*sp.exp(-0.5*(x-y)*D*(x-y).T)
	
#kf=gen_sqexp_k([0.1,0.075,15.,15.,15.,15.])
kf=gen_sqexp_k([1.2,1.3,0.7])
#def kf4(x1,x2):
#    delta=x1-x2
#    d=delta*delta.T
#    return 0.01*sp.exp(-((d[0,0])/0.03))
```

```python
import GPdraw
r=40
ob=GPdraw.gen(kf,[2,2],[-2,-2])
xaxis=sp.linspace(-2,2,r)
yaxis=sp.linspace(-2,2,r)
C=[]
for x in xaxis:
    for y in yaxis:
        C.append(ob(sp.matrix([[x,y]])))
Cgrid=sp.vstack(np.split(np.array(C),r))

plt.contour(xaxis,yaxis,Cgrid,r)
```

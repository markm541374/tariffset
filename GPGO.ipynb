```python
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
import time
plt.ion()
%matplotlib inline
```

```python
def gen_sqexp_k(theta):
	A=theta[0]
	d=sp.matrix(theta[1:]).T
	
	N=sp.matrix(d).shape[0]
	D=sp.eye(N)
	for i in range(N):
		D[i,i]=1./(d[i,0]**2)
	return lambda x,y:A*sp.exp(-0.5*(x-y)*D*(x-y).T)
	
kf=gen_sqexp_k([0.1,0.075,15.,15.,15.,15.])
#def kf4(x1,x2):
#    delta=x1-x2
#    d=delta*delta.T
#    return 0.01*sp.exp(-((d[0,0])/0.03))
```

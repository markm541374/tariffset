#Demo of heating system finding optimum strategy under variable tariff usinf dynamic programing

The heater is going to maintain temperature in the range 15-20 despite external temperature variation. It has a binary output which can switch on every 15 minutes and then remain on for between 0 and all of the 15 minute period.

Temperature is discretised into 100 bins 15 and 20 for dynamic programing.The phyics model is simple exponential cooling, the heater has power 0.0015 and the cooling constant is 0.00005. This implies that it will take 6 hours to fall from 15 above exteral to 5 above external with no input, and at full on the temperature can be maintained 30 above external.

```python
import simple_heating
import numpy as np
from matplotlib import pyplot as plt
%pylab inline
#simple heater that needs to remain within 15 and 20
h = simple_heating.heater([15,20],0.00005,0.0015,15*60,100)

```

```python
#just a function to plot nice graphs, all x axes are time in hours
def ploth(h,cfn,Tfn):
    h.DP(cfn,Tfn)
    h.detail_T()

    print "poweruse:  " + str(h.poweruse)
    print "powercost: " + str(h.day_cost)
    taxis = [i/(60.*60.) for i in h.detT[0]]
    plt.figure(figsize=[10,5])
    plt.plot(taxis,h.detT[1],'r')
    plt.fill_between(taxis,[15.]*len(taxis),[20.]*len(taxis),color='AliceBlue',edgecolor='LightSkyBlue')
    plt.axis([0,24,14,21])
    plt.title("internal temp")
    plt.figure(figsize=[10,5])
    plt.plot(taxis,map(Tfn,h.detT[0]),'r')
    plt.title("external temp")
    plt.figure(figsize=[10,5])
    plt.plot(taxis,h.detT[3],'b')
    plt.plot(taxis,map(cfn,h.detT[0]),'g')
    plt.axis([0,24,0,2])
    plt.title("green:tariff, blue: mark/space ratio")
    return
```

##First check it all looks sensible under constant tariff with daily external temperature oscillation
It does. The temperature steps along the bottom of the allowed range, using only as much power as required to return to 15 every step.

```python
#try out a flat cost and external temp

def flat_cost(t):
    return 1.

def flat_temp(t):
    return 5.

def sine_temp(t):
    return 5.-2*np.cos(np.pi*2*t/(60*60*24))

ploth(h,flat_cost,sine_temp)
```

##Now lets have a late night discount
Power will cost 0.5 between midnight and 7 and 1 otherwise.

The heating goes to max approaching the transition so that temperature peaks at the top of the allowed range at 7am then heating goes to zero and temperature falls back to the lower bound once the tariff has increased. By shifthing the work to a cheaper period the cost has dropped 19% despite using 3% more power

```python
#now lets have late-night discount

def late7_cost(t):
    if t<6.999*60*60:
        return 0.5
    else:
        return 1.

ploth(h,late7_cost,sine_temp)
```
